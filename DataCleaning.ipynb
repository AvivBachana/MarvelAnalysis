### Functions


```python
from rapidfuzz import fuzz, process
import pandas as pd
import numpy as np
import re
from sqlalchemy import create_engine


# Utility Functions
def clean_text(text):
    """
    Clean and standardize text strings.
    """
    if pd.isna(text):
        return ""
    text = str(text).lower().strip()
    text = re.sub(r'[:()\-_.,[\]{}\\/|]', ' ', text)  # Replace special characters with space
    return ' '.join(text.split())  # Remove extra spaces

def extract_number(text):
    """
    Extract numeric sequence indicators from text.
    """
    if pd.isna(text):
        return None
    match = re.search(r'\d+', str(text))
    return int(match.group()) if match else None

def prepare_dataframes(df1, df2, left_on, right_on,how_to="inner"):
    """
    Prepare dataframes for fuzzy matching by adding normalized columns and sequence numbers.
    """
    df1 = df1.copy()
    df2 = df2.copy()
    df1['normalized'] = df1[left_on].apply(clean_text)
    df2['normalized'] = df2[right_on].apply(clean_text)
    df1['number'] = df1['normalized'].apply(extract_number)
    df2['number'] = df2['normalized'].apply(extract_number)
    return df1, df2

def fuzzy_merge(df1, df2, left_on, right_on, number_weight=20, min_score=80,how_to="inner"):

    """
    Optimized fuzzy merge function to match rows based on normalized text similarity and sequence numbers.
    """
    # Preprocess dataframes
    df1_prep, df2_prep = prepare_dataframes(df1, df2, left_on, right_on)
    
    # Fuzzy matching using process.cdist for efficient similarity computation
    matches = process.cdist(
        df1_prep['normalized'].tolist(),
        df2_prep['normalized'].tolist(),
        scorer=fuzz.token_sort_ratio
    )

    # Find the best matches for each row in df1
    best_match_indices = np.argmax(matches, axis=1)
    best_match_scores = np.max(matches, axis=1)

    # Filter matches by minimum score
    match_mask = best_match_scores >= min_score
    matched_indices = np.where(match_mask)[0]

    # Create result dataframe
    results = []
    for i in matched_indices:
        left_row = df1_prep.iloc[i]
        right_row = df2_prep.iloc[best_match_indices[i]]
        score = best_match_scores[i]

        # Adjust score based on number matching
        if left_row['number'] == right_row['number']:
            score += number_weight

        # Append matched rows with scores
        results.append({
            'left_index': left_row.name,
            'right_index': right_row.name,
            'Weighted_Score': score
        })

    # Build final merged dataframe
    match_df = pd.DataFrame(results)
    if match_df.empty:
        return pd.DataFrame()  # Return empty dataframe if no matches

    merged_df = pd.merge(
        df1.reset_index(), 
        match_df, 
        left_index=True, 
        right_on='left_index',
        how= how_to
    )
    merged_df = pd.merge(
        merged_df, 
        df2.reset_index(), 
        left_on='right_index', 
        right_index=True, 
        suffixes=('_df1', '_df2'),
        how= how_to
    )
    merged_df.drop(['left_index', 'right_index'], axis=1, inplace=True)
    return merged_df



def Load_SQL(HOST, USER, PASSWORD, DATABASE, dfs):
    """
    Load data from MySQL database.
    """
    
    # Create a connection to the MySQL database
    engine = create_engine(f"mysql+mysqlconnector://{USER}:{PASSWORD}@{HOST}/{DATABASE}")

    # Example DataFrames (replace these with your actual DataFrames)


    # Load DataFrames to MySQL
    for name, df in dfs.items():
        df.to_sql(name, con=engine, if_exists='replace', index=False)

    print("DataFrames loaded successfully!")
```
